# Filtering

```python
# Filtering
```

```python
df = pd.DataFrame(np.array(([1, 2, 3], [4, 5, 6])),
                index=['ayam', 'tikus'],
                columns=['satu', 'dua', 'tiga'])
df
```

```python
# select columns by name // alternatif lainnya bisa pake metode slicing
df.filter(items=['satu', 'tiga'])
```

```python
df[['satu', 'tiga']]
```

```python
df[['satu','tiga','dua']]
```

```python
df.iloc[0:1,1]
```

```python
# select columns by regular expression
df.filter(regex='a$', axis=1)
```

```python
df.filter(regex='a$', axis=0)
```

```python
# select rows containing 'iku'
df.filter(like='iku', axis=0)
```

```python
df.filter(axis=0, like='a')
```

```python
# select rows containing 'iku'
df2 = df.filter(like='iku', axis=0) |  df.filter(regex='$a', axis =0)
```

```python
df2
```

# Additional Column

```python
# Add new column
```

```python
# Import pandas package
import pandas as pd

# Define a dictionary containing Students data
data = {'Name': ['Andi', 'Budi', 'Citra', 'Dani'],
        'Height': [175, 178, 170, 173],
        'Qualification': ['Msc', 'MA', 'Msc', 'Msc']}

# Convert the dictionary into DataFrame
df = pd.DataFrame(data)
df
```

```python
# Buat sebuah list yang dikonversikan menjadi isi dari kolom
address = ['Jakarta', 'Bogor', 'Bogor', 'Bandung']

# Gunakan 'Address' sebagai nama kolom
# buat persamaanya dengan list
df['Address'] = address
df
```

```python
# Menggunakan DataFrame.insert() untuk add kolom
df.insert(3, "Age", [21, 23, 24, 21], False)
df
```

```python
# Gunakan 'Address' sebagai nama kolom dan buat persamaan list
df2 = df.assign(Address = ['Jakarta', 'Riau', 'Bogor', 'Bandung'])
df2
```

```python
# Mencocokan nilai menggunakan dictionary
address = {'Jakarta': 'Andi', 'Riau': 'Budi',
           'Bogor': 'Citra', 'Bandung': 'Dani'}
```

```python
#Jadikan 'Address' sebagai nama kolom
df['Address'] = address
df
```

```python
df[df.Age > 21] #df['Age']
```

```python
df[(df['Height'] >= 175) & (df.Age >21)]
```

```python
df[df.Name.str.contains('n')]
```
